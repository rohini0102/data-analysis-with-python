**Extracting HTML code of a website:**
**program flow:**
1. import libraries
2. save the url
3. using requests.get() , access the webpage
4. using BeautifulSoup() , access the HTML code

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://en.wikipedia.org/wiki/Crayon_Shin-chan"
```

```python
page=requests.get(url)#access the webpage
```

```python
code=BeautifulSoup(page.text)#to access whole code of webpage
print(code)
```

```python
title=code.find("title")# to extract title or content from webpage
print(title)
```

```python
print(title.text)#text with out tags
```

extracting a table from the webpage

**program flow:**
1. import libraries
2. save the url
3. using requests.get() , access the webpage
4. using BeautifulSoup() , access the HTML code





```python
url2="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
page=requests.get(url2)
```

```python
code=BeautifulSoup(page.text)
print(code)
```

```python
table=code.find("table")# to extract title or content from webpage
print(table)
```

```python
t=table.find_all('th')#to find all the tags in webpage
print(t)
```

```python
headings=[i.text for i in t]#using list comprehension we took only text
print(headings)
```

```python
import pandas as pd
```

```python
rows=table.find_all('tr')
print(rows)
```

```python
data=table.find_all('td')
print(data)
```

```python
data_act=[i.text for i in data]
print(data_act)
```

```python
data_act[0:4]#to get 1st 4
```

```python
dataframe=pd.DataFrame(columns=headings)
print(dataframe)
```

```python
ind=0
st=0
sp=4
for i in data_act:
  if sp<=40:
    dataframe.loc[ind]=data_act[st:sp]
    st=st+4
    sp=sp+4
    ind=ind+1
dataframe
```

```python
dataframe.to_csv("mydf2.csv")
```
