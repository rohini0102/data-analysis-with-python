```python
#write a plotting statement there are 5 subjects psycology,nano technology,medicine,oncology,antrophology 2 students have applied to do masters in any of these 5 feilds eith a line plot show in which feild they can exel. the cut of for every subject is 70.
student1=[67,45,73,71,57]
student2=[43,67,68,53,85]
cut=[70,70,70,70,70]#cutoff
sub=['psycology','nono technology','medicine','oncology','antrophology']
plt.title("subject marks of students")
#plt.plotter type(x,y,label,marker,linewidth,linestyle)
plt.scatter(sub,student1,label="s1",color='red',marker='p')#stdent1 plot
plt.scatter(sub,student2,label="s2",color='pink',marker='H')#stdent2 plot
plt.plot(sub,cut,label="cutoff",color='yellow')
plt.legend()#for labels seen
plt.xlabel('subjects')
plt.ylabel('marks')
```

```python
#an india vs australia match is on live india has give 298 as target for australia the runs for every 5 overs is given as follows:
#india=[25,51,84,131,160,189,220,250,267,297]
#australia have completed only 25 overs
#aus=[15,41,94,110,151]
#plot overs vs runs for both the countries with india as a blue line and austriliya as yellow line .give required lables and legend.
#yhe marker for india is flag ,marker for australia is a plus(+), take overs from a range functions create runs as arrays
#marker=">" [--->flag]
ind=np.array([25,51,84,131,160,189,220,250,267,297])
aus=np.array([15,41,94,110,151])
overs1=np.arange(5,51,5)
overs2=np.arange(5,26,5)
plt.title("cricket score india vs australia")
plt.plot(overs1,ind,label="india",color='b',marker=">",linewidth=2)
plt.plot(overs2,aus,label="australia",color='y',marker="P",linewidth=2)
plt.legend()
plt.xlabel('overs')
plt.xlabel('runs')
plt.grid()#for


```
