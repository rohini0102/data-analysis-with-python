```python
#tuples
```

```python
#to print the elements in tuple
numbers = (1, 2, -5)
print(numbers)
```

```python
#to print the element using index
languages = ('Python', 'Swift', 'C++')
print(languages[0])
print(languages[2])
```

```python
#length of tuple
cars = ('BMW', 'Tesla', 'Ford', 'Toyota')
print('Total Items:', len(cars))
```

```python
#for loop
fruits = ('apple','banana','orange')
for fruit in fruits:
    print(fruit)
```

```python
# tuple including string and integer
mixed_tuple = (2, 'Hello', 'Python')
print(mixed_tuple)
```

```python
#concatination of 2 tuples
fruits = ('apple','banana','orange')
cars = ('BMW', 'Tesla', 'Ford', 'Toyota')
print(fruits+cars)
```

```python
# deleting the tuple
animals = ('dog', 'cat', 'rat')
del animals
```

```python
# creating a tuple from a list
t = tuple([1, 4, 6])
print('t =', t)
```

```python
# creating a tuple from a string
t1 = tuple('Python')
print('t1 =',t1)
```

```python
# creating a tuple from a dictionary
t1 = tuple({1: 'one', 2: 'two'})
print('t1 =',t1)
```

```python
#sets
```

```python
# create set from list
list_numbers = [1, 2, 3, 4, 2, 5]
numbers_set = set(list_numbers)
print(numbers_set)
```

```python
# create a set of integer type
student_id = {112, 114, 116, 118, 115}
print('Student ID:', student_id)

# create a set of string type
vowel_letters = {'a', 'e', 'i', 'o', 'u'}
print('Vowel Letters:', vowel_letters)

# create a set of mixed data types
mixed_set = {'Hello', 101, -2, 'Bye'}
print('Set of mixed data types:', mixed_set)
```

```python
#duplicate items
numbers = {2, 4, 6, 6, 2, 8}
print(numbers)
```

```python
# using add() method
numbers = {21, 34, 54, 12}
print('Initial Set:',numbers)
numbers.add(32)
print('Updated Set:', numbers)
```

```python
# using update() method
companies = {'Lacoste', 'Ralph Lauren'}
tech_companies ={'apple', 'google', 'apple'}
companies.update(tech_companies)
print(companies)
```

```python
# for loop to access each fruits
fruits = {"Apple", "Peach", "Mango"}
for fruit in fruits:
    print(fruit)
```

```python
# find length of set
even_numbers = {2,4,6,8}
print('Set:',even_numbers)
print('Total Elements:', len(even_numbers))
```

```python
#union()
A = {1, 3, 5}
B = {0, 2, 4}
print('Union using |:', A | B)
print('Union using union():', A.union(B))
```

```python
#intersection()
A = {1, 3, 5}
B = {1,2,3}
print('intersection using &:', A & B)
print('intersection using intersection():', A.intersection(B))
```

```python
#difference()
A = {1, 3, 5}
B = {1,2,3}
print('difference using -:', A - B)
print('difference using difference():', A.difference(B))
```
